```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.utils import shuffle
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# === КРОК 1: Завантажити базу параметрів квітів ===
df = pd.read_csv('IrisData_full.csv')

print("Перші 5 рядків завантаженої бази:")
print(df.head())

# === КРОК 2: Перемішати записи ===
df = shuffle(df, random_state=42).reset_index(drop=True)

# Розділяємо на ознаки (X) та класи (y)
X = df.iloc[:, :-1].values # Всі колонки крім останньої
y = df.iloc[:, -1].values  # Остання колонка (назви квітів)

# === КРОК 3: Нормалізувати параметри ===
# KNN чутливий до масштабу даних, тому це важливо
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# === КРОК 4: Розділити на навчальну і тестову вибірки ===
# test_size=0.3 означає, що 30% піде на тест. stratify=y зберігає баланс класів.
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.3, random_state=42, stratify=y
)

print(f"\nРозмір навчальної вибірки: {X_train.shape}")
print(f"Розмір тестової вибірки: {X_test.shape}")

# === КРОК 5: Навчити KNN з різними К (від 1 до 20) ===
k_values = range(1, 21)
acc_train = []
acc_test = []

for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    
    acc_train.append(accuracy_score(y_train, knn.predict(X_train)))
    acc_test.append(accuracy_score(y_test, knn.predict(X_test)))

# === КРОК 6: Вибрати найкраще К ===
best_index = np.argmax(acc_test) 
best_k = k_values[best_index]
best_acc = acc_test[best_index]

print(f"\nНайкраще значення K: {best_k}")
print(f"Найкраща точність (Accuracy): {best_acc}")

# === КРОК 7: Перевірка результату та Графік ===

best_knn = KNeighborsClassifier(n_neighbors=best_k)
best_knn.fit(X_train, y_train)
y_pred = best_knn.predict(X_test)

print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))
print("\nClassification Report:")
print(classification_report(y_test, y_pred))


plt.figure(figsize=(10, 6))
plt.plot(k_values, acc_train, label='Train Accuracy', marker='o')
plt.plot(k_values, acc_test, label='Test Accuracy', marker='o')
plt.xlabel('k (кількість сусідів)')
plt.ylabel('Accuracy (Точність)')
plt.title('Залежність точності від k')
plt.xticks(k_values)
plt.legend()
plt.grid(True)
plt.show()
```
